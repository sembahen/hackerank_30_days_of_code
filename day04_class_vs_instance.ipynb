```
class Pounds:
  def __init__(self,initialweight):
    if initialweight < 0:
      print("What planet are you from?")
      self.weight=5
    else:
      self.weight=initialweight
  
  def ami_obese(self):
    if self.weight<80:
      print("You are light")
    elif self.weight<175:
      print("You are average")
    else:
      print("You are heavy.")
   
  def exercise(self):
    self.weight-=3

w=int(input())
for i in range(0,w):
  weight=int(input())
  p=Pounds(weight)
  p.ami_obese()
  for j in range(0,3):
    p.exercise()
  p.ami_obese()
      
```
