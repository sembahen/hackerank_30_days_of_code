```
n=int(input())
def consecutive_ones(n):
  current_consec_ones=0
  max_consec_ones=0
  while n>0:
    if n%2==1:
      current_consec_ones+=1
      if current_consec_ones > max_consec_ones:
        max_consec_ones=current_consec_ones
    else:
      current_consec_ones=0
    n=n//2
  return max_consec_ones
consecutive_ones(n)
```
